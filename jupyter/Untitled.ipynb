```python
import mss
import numpy as np
import matplotlib.pyplot as plt
import pyautogui
```

```python
def screenshot(region):    
    with mss.mss() as sct:
        return np.array(sct.grab(region))
```

```python
def clean_img(img):
    img[img < 128] = 0
    img[img >= 128] = 1
    return img
```

```python
def processed_img(region):
    img = screenshot(region)
    img = img[:, :, 0]
    img = clean_img(img)
    return img
```

```python
top, left = 100, 100
```

```python
width, height = 600, 600
```

```python
s_region = { 'top': 100, 'left': 10, 'width': width, 'height': height }
```

```python
img = processed_img(s_region)
```

```python
%matplotlib qt
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```
